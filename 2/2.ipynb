```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0, 5, 11)
y = x ** 3
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title') 
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 2, 2])

axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Title')

axes2.plot(x, y, 'b')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Title')
```

```python
fig,axes = plt.subplots()

axes.plot(x, y, 'b')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes = plt.subplots(nrows=4,ncols=4)
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x, x, label="x")
ax.plot(x, x**2, label="x**2")
ax.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x, color = 'g')
ax.plot(x, x**2, color = 'b')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, lw = 4, ls = '--')
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, marker = 'o', markersize = 10, markerfacecolor = 'y')
```
